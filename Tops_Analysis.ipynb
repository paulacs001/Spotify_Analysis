# Calculating Similarity

Once we have our dataset in a suitbale form, we must determine a single playlist vector, and compare with the other songs to see which are most similar.

This will be done using cosine similarity. The lower the score, the better!

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

#want to find most similar country to Spain
c1 = pd.DataFrame(country_df.loc["España"])
c1 = np.transpose(c1)

res = cosine_similarity(country_df, c1)


res = list(map(lambda x: x[0], res))

sorted = [i for i in res]
sorted.sort()

#The one with least distance is itself, so omit
top_5 = sorted[-6:-1]

ids = []
for i in top_5:
  id = np.where(res == i)
  ids.append(id[0][0])


country_df.iloc[ids,:]
```

### Similarity Dataframe


```python
#For our visualizations, we want to have a matrix with the similarity between all countries
from sklearn.model_selection import ParameterGrid
param_grid = ParameterGrid({'Origin': countries, 'Dest': countries, 'Similarity': [0]})

country_similarity_df = pd.DataFrame(list(param_grid))

for i in range(len(country_similarity_df)):
  c1 = pd.DataFrame(country_df.loc[country_similarity_df.loc[i, 'Origin']]).transpose()
  c2 = pd.DataFrame(country_df.loc[country_similarity_df.loc[i, 'Dest']]).transpose()
  country_similarity_df.loc[i, 'Similarity'] = cosine_similarity(c1,c2)[0]

country_similarity_df
```

# Recommending songs for each country

Now, for each country we will be recommending 25 new songs that could possibly fit the countries musical tastes. For this, we will be using both songs from other Top50 playlists, and that info collected from public playlists (removing info concerning mood).

We will perform the same transformations that in the latter dataset; extract the release year and normalize numerical values.

Also, we will remove duplicates!


>Programatically, we find the ids of the recommended songs, and from here we will create the database with its Title and Artists for easier visualization.

### Getting recommended song ids

```python
### ------------MOODS------------ ###
data_moods = pd.read_csv('song_feat_moods.csv').drop('Unnamed: 0',axis =1)

#Extract year
data_moods['release_year'] = data_moods['release_date'].map(lambda x: x.split('-')[0]).astype(int)
data_moods.drop('release_date', axis = 1, inplace = True)

#Normalize
num = data_moods.select_dtypes(['float64', 'int64']).columns.tolist()
num_data = data_moods[num]

scaler = StandardScaler()
num_data = scaler.fit_transform(num_data)

num_data = pd.DataFrame(num_data, columns = num)

#Combine
data_moods = pd.concat([num_data, data_moods['id']], axis = 1)


### ------------TOPS------------ ###
data_tops = data = pd.read_csv('song_feat_top.csv').drop('Unnamed: 0',axis =1)

#Extract year
data_tops['release_year'] = data_tops['release_date'].map(lambda x: x.split('-')[0]).astype(int)
data_tops.drop('release_date', axis = 1, inplace = True)

#Normalize
num = data_tops.select_dtypes(['float64', 'int64']).columns.tolist()
num_data = data_tops[num]

scaler = StandardScaler()
num_data = scaler.fit_transform(num_data)

num_data = pd.DataFrame(num_data, columns = num)

#Combine
data_tops = pd.concat([num_data, data_tops['country'], data_tops['id']], axis = 1)


### ------------ ALL ------------ ###
data_tot = pd.concat([data_tops, data_moods], axis = 0)
data_tot = data_tot.drop_duplicates(subset='id', keep="last")
data_tot.iloc[:,:-2]
# data_tot.drop('id', axis = 1)
```

```python
# For each country: 
#     - recall its country playlist vector
#     - filter data_tot as to not include songs from that country
#     - compute similarity and keep only 25 best

rec_per_country = {}
sim_per_country =  []
for c in countries:
  #Retrieve data
  country_vector = pd.DataFrame(country_df.loc[c]).transpose()
  tmp = data_tot[data_tot.country != c]

  #Compute similarities and sort them
  res = list(cosine_similarity(country_vector, tmp.iloc[:,:-2])[0])

  sorted = [i for i in res]
  sorted.sort()

  #Retrieve top 25
  top_25 = sorted[-25:]

  ids = []
  for i in top_25:
    id = np.where(res == i)
    ids.append(id[0][0])

  #Store list with country's 25 recommended songs' ids
  rec_per_country[c] = list(tmp.iloc[ids,-1]) 

  #Store the similarity values for those recommended songs
  sim_per_country.append(top_25)

sim_per_country = list(itertools.chain(*sim_per_country))
```

```python
def parse_playlist(playlist):
    playlist_dict = {}
    print(playlist[0])
    for i in playlist:
        song = i['track']['name']
        id_ = i['track']['id']
        artists = i['track']['artists']
        
        #playlist_dict[id_] = {'name': song}
        playlist_dict[id_] = {'name': song, 'artists': artists}  # necesario para lyrics en genius
    return playlist_dict
```

```python
def retrieve_songs(playlist_id):
    """
    retrieve the information of the songs of the playlists
    """

    song_ids = []
    url = f'https://api.spotify.com/v1/playlists/{playlist_id}/tracks?fields=items(track(name,id,artists(name)))'
    response = requests.get(url,
        headers = {
            'Authorization': f'Bearer {AUTH_TOKEN}'
        }        
    )
    
    json_response = response.json()

    res = json_response['items']            
    for song in res:
      try:
        song_ids.append(song['track'])
      except:
        pass

    return song_ids
```
